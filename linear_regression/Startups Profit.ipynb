```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Data/Startups.csv')
```

```python
df.head()
```

```python
encoder = LabelEncoder()
df['numerical_State'] = encoder.fit_transform(df['State'])
df.drop('State', inplace=True, axis=1)
df.rename(columns={'R&D Spend': 'Research and Development', 'Marketing Spend': 'Marketing', 'numerical_State': 'State'}, inplace=True)
```

```python
df.head()
```

```python
correlations = df.corr()
sns.heatmap(correlations, cmap='coolwarm')
```

```python
scaler = StandardScaler()
df_standardised = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
X_features = ['Research and Development', 'Administration', 'State']
X = df[X_features]
Y = df.Profit
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, Y_train)
```

```python
Y_pred = lr_model.predict(X_test)
mae = mean_absolute_error(Y_test, Y_pred)
mae
```
